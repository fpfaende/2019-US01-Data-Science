```python
import pandas as pd
```

```python
df = pd.read_pickle('savedDF.pickle')
df.head()
```

```python
bti_df = pd.read_csv('data-BTI-2006-2018.csv')
bti_df[bti_df['year'] == 2018]
bti_df[bti_df['country_code'] == 'BRN']
```

```python
bti_2018 = bti_df[bti_df['year'] == 2018]
for index, isocode in df.isocode.iteritems():
    try:
        ind = bti_2018[bti_2018['country_code']==isocode]['stat_ind']
        rank = bti_2018[bti_2018['country_code']==isocode]['rank_stat_ind']
        df.at[index,'Bertelsmann Transformation index (BTI)'] = ind
        df.at[index,'Bertelsmann Transformation index (BTI) - Ranking'] = rank
    except ValueError:
        print('[critical]:no data for',isocode)
        continue
df[['Country (according to CN Gov):','Bertelsmann Transformation index (BTI)','Bertelsmann Transformation index (BTI) - Ranking' ]]
```

```python
cpi_df = pd.read_excel('data-CPI-2018.xlsx', header=2)
cpi_df.head()
cpi_df.columns
```

```python
for index, isocode in df.isocode.iteritems():
    try:
        ind = cpi_df[cpi_df['ISO3']==isocode]['CPI Score 2018']
        rank = cpi_df[cpi_df['ISO3']==isocode]['Rank ']
        df.at[index,'TI Corruption Perception Index'] = ind
        df.at[index,'TI Corruption Perception Index - Ranking'] = rank
    except ValueError:
        print('[critical]:no data for',isocode)
        continue
```

```python
heritage_df = pd.read_excel('data-heritage-2019.xls')
```

```python
for index, name in df['Country (according to CN Gov):'].iteritems():
    cleaned_name = name.split('(')[0].strip()
    if cleaned_name == 'Thailiand':
        cleaned_name = 'Thailand'
    elif cleaned_name == 'The United Arab Emirates':
        cleaned_name = 'United Arab Emirates'
    elif cleaned_name == 'Brunei':
        cleaned_name = 'Brunei Darussalam'
    elif cleaned_name == 'Myanmar':
        cleaned_name = 'Burma'
    elif cleaned_name == 'East Timor':
        cleaned_name = 'Timor-Leste'
    elif cleaned_name == 'Kyrgyzstan':
        cleaned_name = 'Kyrgyz Republic'
    elif cleaned_name == 'United States of America':
        cleaned_name = 'United States'
    
    
    try:
        ind = heritage_df[heritage_df['Country Name']==cleaned_name]['2019 Score']
        rank = heritage_df[heritage_df['Country Name']==cleaned_name]['World Rank']
        df.at[index,'HF Index of Economic Freedom (Scale: 1-100)'] = ind
        df.at[index,'Heritage Foundation Index of Economic Freedom (Global Ranking)'] = rank
    except ValueError:
        print('[critical]:no data for',cleaned_name)
        continue
        
```
